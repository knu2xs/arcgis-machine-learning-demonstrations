```python
import arcpy
import pandas as pd
import os
```

```python
arcpy.env.workspace = path.join(os.getcwd(), r'resources\data.gdb')
stores = 'stores_all'
```

```python
field_names = ['LOCNUM', 'CONAME']
df = pd.DataFrame([_ for _ in arcpy.da.SearchCursor(stores, field_names)], columns=field_names)
df.head()
```

```python
target_count = df.CONAME.value_counts()['TARGET']
target_count
```

```python
df_sample = df.sample(n=target_count)
df_sample.head()
```

```python
where_clause = ' OR '.join(["LOCNUM = '{}'".format(locnum) for locnum in df_sample['LOCNUM']])
where_clause
```

```python
out_fc = 'stores_sample'
if arcpy.Exists(out_fc):
    arcpy.Delete_management(out_fc)
arcpy.Select_analysis(stores, out_fc, where_clause)[0]
```
